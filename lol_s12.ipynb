```python
df = pd.read_csv('lol-s12.csv')
```

```python
df.head()
```

```python
df["Picks"] = df["Picks"].astype(int)
df["Presence"] = df["Presence"].astype(int)
df["Winrate"] = df["Winrate"].astype(float)
df = df.loc[df["KDA"] != '-']
df["KDA"] = df["KDA"].astype(float)

df = df[df["Presence"]>0]
df = df.dropna()
```

```python
def plot_scatter(df, target_columns):
    df2 = df[["Champion"]+target_columns]

    plt.figure(figsize=(10, 10))
    x = df[target_columns[0]]
    y = df[target_columns[1]]
    plt.scatter(x, y)

    ax = plt.gca()
    for i, txt in enumerate(df["Champion"]):
        ax.annotate(txt, (x[i]+1, y[i]+1))

def plot_scatter2(df, target_columns, size=None, pos=None, horizon=50, color=None):
    if pos:
        df = df[df['Pos']==pos]
    import plotly.express as px
    x = target_columns[0]
    y = target_columns[1]

    fig = px.scatter(df, x=x, y=y, text="Champion", log_x=False,# size_max=100,
                     size=[i+1 for i in df[str(size)].tolist()] if size else None,
                     color=color if color else None,
                     color_continuous_scale=px.colors.sequential.YlOrRd,
                     width=800, height=600)
    fig.update_traces(textposition='top center')
    if horizon:
        fig.add_hline(y=horizon, line_width=2, line_dash='dash', line_color="red")
    fig.update_layout(title_text=x + "  v.s. " + y if not pos else f"{pos}: " + x + "  v.s. " + y, 
                      title_x=0.5)
    fig.show()
    return df
    
# plot_scatter(df, ["Presence", "Winrate"])
plot_scatter2(df, ["Presence", "Winrate"])
```
